```python
import numpy as np
import tensorflow as tf
```

## Generate data

```python
xs = np.array([-1.0, 0.0, 1.0, 3.0, 4.0], dtype = float)
ys = 2*xs-1
ys
```

## Define a neural network

```python
model = tf.keras.Sequential([
    #input layer
    tf.keras.Input(shape=(1,)),
    #dense layer, linear layer
    tf.keras.layers.Dense(units=1)
])
```

- `Sequential Class` allows you to define the network as ***sequence of layers***.
- `Dense` is a linear layer.
- You define shape of input to network in Input layer.


```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(xs, ys, epochs = 500)
```

```python
model.predict(np.array([10.0]), verbose=1)
```
